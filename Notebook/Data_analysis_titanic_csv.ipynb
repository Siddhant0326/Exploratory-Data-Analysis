```python
#Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the dataset
df = pd.read_csv("titanic.csv")

df.head()
```

```python
#To check no. of rows and columns
df.shape
```

```python
#To check column names datatypes
df.info()
```

```python
#To check the basic statistics
df.describe()
```

```python
#To check missing values
df.isnull().sum()
```

```python
#Dropping the cabin column as it contains too many missing values
df = df.drop("Cabin", axis=1)
```

```python
#Imputting the missing values in the column age with the mean of column
df["Age"].fillna(df["Age"].mean())
```

```python
#Handling missing values in column embarked using forwardfilling method
df["Embarked"].fillna(method="ffill")
```

```python
#Visualization
```

```python
#Histogram
sns.histplot(df['Age'], bins=20, kde=True)
plt.show()
```

```python
#Bar plot
df['Sex'].value_counts().plot(kind="bar")

plt.xticks(rotation=0)   
plt.show()
```

```python
#Scatter Plot
plt.scatter(df["Age"], df["Fare"])
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
#Pie Chart   0 = Not Survived, 1 = Survived
df["Survived"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.ylabel("")   
plt.title("Pie Chart")
plt.show()
```
