```python
#Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the dataset
df = pd.read_csv("iris.csv")

df.head()
```

```python
#To check no. of columns and rows
df.shape
```

```python
#To check column names datatypes
df.info()
```

```python
df.isnull().sum()
```

```python
#Pair Plot
sns.pairplot(df)
plt.show()
```

```python
#Boxplot
sns.boxplot(data=df)
plt.show()
```

```python
#Heatmap correlation
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```

```python
#Violin plot
sns.violinplot(x=df["species"], y=df["petal_width"])
plt.show()
```
